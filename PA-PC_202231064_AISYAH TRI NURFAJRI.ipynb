```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = "PA-PC_202231064_AISYAH TRI NURFAJRI_B.jpg"
original_image = cv2.imread(image_path)
original_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)
```

```python
(h, w) = original_image.shape[:2]
center = (w / 2, h / 2)
M = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated_image = cv2.warpAffine(original_image, M, (w, h))
```

```python
resized_image = cv2.resize(original_image, (50, 100))
```

```python
cropped_image = original_image[200:370, 100:250]
```

```python
flipped_image = cv2.flip(original_image, 1)
```

```python
M = np.float32([[1, 0, 50], [0, 1, 50]])
translated_image = cv2.warpAffine(original_image, M, (w, h))
```

```python
titles = ['Citra Asli', 'After Rotated', 'After Resized', 'After Cropped', 'After Flipped', 'After Translated']
images = [original_image, rotated_image, resized_image, cropped_image, flipped_image, translated_image]

plt.figure(figsize=(12, 8))

for i in range(6):
    plt.subplot(2, 3, i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('on')

plt.tight_layout()
plt.show()
```
